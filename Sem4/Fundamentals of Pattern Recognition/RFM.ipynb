```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
df = pd.read_excel('Online Retail.xlsx')
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Total_price'] = df['Quantity']*df['UnitPrice']
```

```python
df.drop('Description',axis=1,inplace=True)
```

```python
df.head()
```

```python
df_rfm = pd.DataFrame()
```

```python
mon = df.groupby('CustomerID')['Total_price'].sum()
```

```python
mon
```

```python
df_rfm['CustID'] = mon.index
```

```python
df_rfm['mon'] = mon.values
```
