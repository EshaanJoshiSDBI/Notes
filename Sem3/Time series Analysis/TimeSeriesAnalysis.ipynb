1. No trend - increasing or decreasnig
2. Mean is constant over a period time
2. Variance of std deviation is constant over a period time

`It is a stationary time series`

````r
skirts = read_excel('skirts.xlsx')
````

````r
skirt_ts = ts(read_excel('skirts.xlsx'))
````

````r
plot(skirt_ts,las = 1)
````

````r
diff_skirts = c(9,8,11)
````

````r
for (i in 1:length(skirt_ts)-1){
    diff_skirts[i] = abs(skirt_ts[i] - skirt_ts[i+1])
}
````

````r
diff_skirts
````

````r
length(skirt_ts) == length(diff_skirts)
````

````r
diff1 = diff(skirt_ts,differences = 1)
````

````r
diff1
````

````r
diff2 = diff(skirt_ts,differences = 2)
diff3 = diff(skirt_ts,differences = 3)
diff4 = diff(skirt_ts,differences = 4)
diff5 = diff(skirt_ts,differences = 5)
diff6 = diff(skirt_ts,differences = 6)
````

````r
cbind(skirt_ts,diff1,diff2,diff3,diff4,diff5)
````

````r
skirt_ts[1]
````

````r
par(mfrow = c(3,3))
plot(skirt_ts)
plot(diff1)
plot(diff2)
plot(diff3)
plot(diff4)
plot(diff5)
plot(diff6)
````
