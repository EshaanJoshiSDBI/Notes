```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
table = pd.pivot_table(data = df, index = ['sex'],values = ['total_bill'])
table
```

```python
table1 = pd.pivot_table(data = df, index = ['sex','smoker'],values = ['total_bill'])
table1
```

```python
table2 = pd.pivot_table(data = df, index = ['sex'])
table2
```

```python
df.head()
```

```python
table.plot(kind = 'bar')
```

```python
table3 = pd.pivot_table(df,index = ['sex','smoker'])
table3
```

```python
table4 = pd.pivot_table(df,index = ['sex','smoker'], aggfunc = {'total_bill': 'mean', 'size': 'sum'})
table4
```

```python
table5 = pd.pivot_table(df, index = ['sex','smoker'], values = ['total_bill'], aggfunc = np.mean)
table5
```

```python
table5.plot(kind = 'bar')
```

```python
table6 = pd.pivot_table(df, index = ['sex'], columns = ['smoker'], values = ['total_bill'], aggfunc = np.sum)
table6
```

```python
table6.plot(kind = 'bar')
```

```python
table7 = pd.pivot_table(df, index = ['sex','smoker','time'], values = ['total_bill'], columns = ['day'], aggfunc = np.mean)
table7
```

```python
table7.plot(kind = 'bar')
```

```python
# handling null values
table10 = pd.pivot_table(df, index = ['sex','smoker','time'], columns = ['day'], values = ['total_bill'], aggfunc = np.mean, fill_value = 0)
table10
```

```python
# handling null values
table11_1 = pd.pivot_table(df, index = ['sex','smoker','day'], columns = ['time'], values = ['total_bill','size'], aggfunc = {'total_bill':'mean', 'size':'sum'}, fill_value = 0)
table11_1
```
